```python
import pandas as pd
import numpy as np
import requests
from tqdm import tqdm

url = "https://cdaw.gsfc.nasa.gov/CME_list/UNIVERSAL_ver1/text_ver/univ_all.txt"

response = requests.get(url)
lines = response.text.split("\n")[4:-2]  # split the data into lines and remove the header and footer lines

# Prepare an empty list to hold the parsed data
data = []

# Parse each line
for line in tqdm(lines, desc="Processing data"):
    row = line.split()
    date = row[0]
    time = row[1]
    central_pa = row[2]
    width = row[3]
    linear_speed = row[4]
    second_order_speed_initial = row[5]
    second_order_speed_final = row[6]
    speed_at_20R = row[7]
    acceleration = row[8]
    mass = row[9]
    kinetic_energy = row[10]
    mpa = row[11]
    remarks = " ".join(row[12:])
    data.append([date, time, central_pa, width, linear_speed, second_order_speed_initial,
                 second_order_speed_final, speed_at_20R, acceleration, mass, kinetic_energy, mpa, remarks])

# Convert the list to a DataFrame
cme_data = pd.DataFrame(data, columns=['Date', 'Time', 'Central_PA', 'Width', 'Linear_Speed', 
                                       '2nd_order_speed_initial', '2nd_order_speed_final', '20R_speed', 
                                       'Accel', 'Mass', 'Kinetic_Energy', 'MPA', 'Remarks'])

# Save to CSV
save_path = "D:\\_NILESH\\dissertation\\data\\cme.csv"
cme_data.to_csv(save_path, index=False)
print(f"Data saved to {save_path}")
```

```python
from sunpy.net import hek
from sunpy.net import attrs as a
import pandas as pd
from tqdm import tqdm
from datetime import timedelta, datetime

# Initialize the HEK client
client = hek.HEKClient()

# Define the time range for the search
tstart = '1986/01/01 00:00:00'
tend = '2019/12/31 00:00:00'
tstart_datetime = datetime.strptime(tstart, '%Y/%m/%d %H:%M:%S')
tend_datetime = datetime.strptime(tend, '%Y/%m/%d %H:%M:%S')

# Prepare an empty dataframe to store results
df = pd.DataFrame(columns=['Date_and_Time', 'X_Coordinates', 'Y_Coordinates'])

# Create a time range from start to end date
time_range = pd.date_range(tstart_datetime, tend_datetime, freq='H')

# Use tqdm for a progress bar
for time in tqdm(time_range):
    # Convert time to string
    time_str = time.strftime('%Y/%m/%d %H:%M:%S')
    next_hour_str = (time + timedelta(hours=1)).strftime('%Y/%m/%d %H:%M:%S')

    # Query the HEK for sunspot data
    result = client.search(a.Time(time_str, next_hour_str), a.hek.EventType('SS'))

    # Add each sunspot event to the dataframe
    for item in result:
        df = df.append({'Date_and_Time': item['event_starttime'].datetime,
                        'X_Coordinates': item['hpc_x'],
                        'Y_Coordinates': item['hpc_y']}, ignore_index=True)
df.to_csv(data_dir + 'sunspots.csv', index=False)
```

```python
# Code from part 2 to get the cycle information
import pandas as pd
from scipy.signal import argrelextrema
import numpy as np

# Load the data
url = "http://www.sidc.be/silso/INFO/snmtotcsv.php"
suns_data = pd.read_csv(url, delimiter=';', header=None)

# Rename the columns
suns_data.columns = ['Year', 'Month','Date in fraction of year', 'Monthly mean total sunspot number', 
                'Monthly mean standard deviation', 'Number of observations', 'Definitive/Provisional marker']

suns_data['Date'] = suns_data['Date in fraction of year'].apply(convert_fractional_year)

# Convert the 'Actual Date' to the format 'DD-MM-YYYY'
suns_data['Date'] = suns_data['Date'].dt.strftime('%d-%m-%Y')

# Convert the 'Date' column to datetime
suns_data['Date'] = pd.to_datetime(suns_data['Date'], format='%d-%m-%Y')

# Set 'Date' as the index
suns_data.set_index('Date', inplace=True)

# Resample the data to get the yearly mean sunspot count
yearly_data = suns_data['Monthly mean total sunspot number'].resample('Y').mean()

# Find local minima
local_min = argrelextrema(yearly_data.values, np.less)[0]

# Prepare an empty dataframe to store cycle data
cycles = pd.DataFrame(columns=['Cycle_Number', 'Cycle_Start_Date', 'Cycle_End_Date', 'Number_of_Days'])

# Compute cycle lengths and store in the dataframe
for i in range(1, len(local_min)):
    start = yearly_data.index[local_min[i-1]]
    end = yearly_data.index[local_min[i]]
    length = end - start
    cycles = cycles.append({'Cycle_Number': i, 
                            'Cycle_Start_Date': start, 
                            'Cycle_End_Date': end, 
                            'Number_of_Days': length.days}, ignore_index=True)

# Code from part 1 to get the sunspot event data
df = pd.read_csv(data_dir + 'sunspots.csv')

# Convert 'Date_and_Time' to datetime
df['Date_and_Time'] = pd.to_datetime(df['Date_and_Time'])

# Merge the cycle data with the sunspot events data
df = pd.merge_asof(df.sort_values('Date_and_Time'), 
                   cycles.sort_values('Cycle_Start_Date'), 
                   left_on='Date_and_Time', 
                   right_on='Cycle_Start_Date', 
                   direction='backward')

# Save the merged dataframe as a CSV file
df.to_csv(data_dir + 'sunspots_with_cycles.csv', index=False)
```

```python
from sunpy.net import hek
from sunpy.net import attrs as a
from datetime import datetime, timedelta
import pandas as pd
from tqdm import tqdm

# Initialize the HEK client
client = hek.HEKClient()

# Prepare an empty dataframe to store results
df = pd.DataFrame()

# Define the start and end years
start_year = 1980
end_year = datetime.now().year

# Use tqdm for a progress bar
for year in tqdm(range(start_year, end_year+1)):
    # Define the start and end times for this year
    tstart = f'{year}/01/01 00:00:00'
    tend = f'{year}/12/31 23:59:59'
    
    # Query the HEK for flare data
    result = client.search(a.Time(tstart, tend), a.hek.EventType('FL'))
    
    # Append the result to the dataframe
    df = df.append(pd.DataFrame(result))

# Save the dataframe as a CSV file
# data_dir = "your/directory/path/"
df.to_csv(data_dir + 'solar_flare_new.csv', index=False)
```
