```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample a subset of the data for visualization
sample_data = solar_flare_energy_data.sample(n=10000, random_state=42)

# List of columns to plot
columns_to_plot = ['energy_e', 'energy_p']

plt.figure(figsize=(15, 10))

for i, col in enumerate(columns_to_plot, 1):
    plt.subplot(1, 2, i)
    sns.histplot(sample_data[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# List of columns to compute distribution values
columns_to_analyze = ['energy_e', 'energy_p']

# Compute distribution values
distribution_values = {}
for col in columns_to_analyze:
    distribution_values[col] = {
        'Mean': solar_flare_energy_data[col].mean(),
        'Median': solar_flare_energy_data[col].median(),
        'Mode': solar_flare_energy_data[col].mode()[0],
        'Variance': solar_flare_energy_data[col].var(),
        'Standard Deviation': solar_flare_energy_data[col].std(),
        'Skewness': solar_flare_energy_data[col].skew(),
        'Kurtosis': solar_flare_energy_data[col].kurt()
    }

distribution_values
```

```python
correlation_matrix = solar_flare_energy_data[['energy_e', 'energy_p']].corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix for Energy Columns")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the dataset with high float precision
solar_flare_energy_df = pd.read_csv('D:\\_NILESH\\dissertation\\data\\cleaned\\solar_flare_energy.csv', float_precision='high')

# 2. Summing the energies (before subsetting the columns)
solar_flare_energy_df['energy_e'] = solar_flare_energy_df.filter(like='electron').sum(axis=1)
solar_flare_energy_df['energy_p'] = solar_flare_energy_df.filter(like='proton').sum(axis=1)

# 3. Subset the desired columns
solar_flare_energy_df = solar_flare_energy_df[['Time','energy_e','energy_p','energy_e_38_53','energy_e_175_315','energy_p_47_68','energy_p_115_195','energy_p_310_580','energy_p_795_1193','energy_p_1060_1090']]

# Convert the "Time" column to datetime format and set as 'date'
solar_flare_energy_df["date"] = pd.to_datetime(solar_flare_energy_df["Time"])

# 4. Data Cleaning: Replace predefined missing values (-1.00E+08) with NaN
columns_to_clean = solar_flare_energy_df.columns.difference(['Time', 'date'])  # Exclude Time and date columns
for column in columns_to_clean:
    solar_flare_energy_df[column].replace(-1.00E+08, value=pd.NA, inplace=True)

# 5. Plotting using a random sample
sample_energy_df = solar_flare_energy_df.sample(frac=0.01, random_state=42)
plt.figure(figsize=(14, 6))
plt.plot_date(sample_energy_df['date'], sample_energy_df['energy_e'], linestyle='solid', marker=None, color='blue', label="Electron Energy")
plt.plot_date(sample_energy_df['date'], sample_energy_df['energy_p'], linestyle='solid', marker=None, color='red', label="Proton Energy")
plt.title('Solar Flare Energy Over Time (Sampled Data)')
plt.xlabel('Date')
plt.ylabel('Energy (particles/cm2-s-ster-KeV)')
plt.legend()
plt.tight_layout()
plt.show()
```
